```python
class list_operations:
    def create_list(self,n,l):
        self.n=n
        self.l=l
        for i in range(self.n):
            self.l.append(input('Enter element:'))
        return self.l
    
    def insert_item(self,item,pos,list1):
        self.item=item
        self.pos=pos
        self.list1=list1
        self.list1.insert(self.pos,self.item)
        return self.list1
    
    def remove_item(self,item,list1):
        self.item=item
        self.list1.remove(item)
        return self.list1
    
    def pop_item(self,pos,list1):
        self.pos=pos
        self.list1=list1
        self.list1.pop(self.pos)
        return self.list1
    
    def delete_list(self,list1):
        self.list1=list1
        self.list1.clear()
        return self.list1
    
    def show_list(self,list1):
        self.list1=list1
        return self.list1
```

```python
def main():
    cmd = 'Start'
    c1 = list_operations()
    l=[]
    while cmd.upper()!='N':
        opr = int(input('''Operations you can perform :
                        1. Create list
                        2. Insert element
                        3. Remove element
                        4. Delete list
                        5. Show list :'''))
        if opr==1:
            l = c1.create_list(int(input('Enter length of list you want to create:')),l)
            print('List created :')
        elif opr==2:
            if (input('Do you want to enter element at particular position? ').upper()=='YES'):
                l = c1.insert_item(input('Enter element:'),int(input('Enter position:')),l)
            else:
                l=c1.insert_item(input('Enter element'),len(l)+1,l)
            print('Element inserted')
        elif opr==3:
            if input('Do you want to remove particular element ? ').upper()=='YES':
                try:
                    l = c1.remove_item(input('Enter element you want to remove:'),l)
                    print('Element removed')
                except:
                    print('Element does not exists in list')
            else:
                try:
                    l = c1.pop_item(int(input('Enter position of element :')),l)
                    print('Element removed..')
                except:
                    print('Enter valid position..')
        elif opr==4:
            l = c1.delete_list(l)
            print('List deleted..')
        elif opr==5:
            print(c1.show_list(l))
        else:
            print('Wrong choice!')
        cmd = input('Do you want to continue? Enter Y else N :')
```

```python
if __name__=='__main__':
    main()
```
