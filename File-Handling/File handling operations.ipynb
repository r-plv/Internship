```python
# Command to enter in While loop
cmd = 'Start'
# Checking condition of while loop
while cmd.upper()!='N':
    # Opening file in read mode and creating object
    with open('F:/Pallavi/CodeSpyder/Internship/Internship_Prep.txt') as file:
        opr = int(input('''Operations you can perfrom -
                        1 Read 1st N lines \n 2 Read Last N lines 
                        3 find Largest word \n 4:Count Characters:'''))
        # Checking condition
        if opr==1:
            no = int(input('How many lines you want to read?:'))
            # Reading lines of file and storing in list (using readlines() method)
            lines = file.readlines()
            # Using loop to read specific no of lines
            for textline in (lines[:no]):
                # printing read lines
                print(textline,end=' ')
        elif opr==2:
            no = int(input('How many lines you want to read?:'))
            lines = file.readlines()
            # using -ve index to read last no of lines
            for textline in (lines[-no:]):
                    print(textline,end=' ')
        elif opr==3:
            # spliting file with space and storing words in list
            words = file.read().split()
            # counting len of every word and storing maximum length
            maxlen = len(max(words,key=len))
            # checking word length with maximum length & 
            # printing word with maximum length using list comprehension
            print('Word with largest length are',[word for word in words if len(word)==maxlen])
        elif opr==4:
            count = 0
            # spliting file with space and storing words in list
            words = file.read().split()
            # loop for counting length of each word
            for character in words:
                count+=len(character)
            print('Characters in file are',count)
        else:
            print('Trying to perform UNAVAILABLE Operation!')
    # Decide to continue while loop or not
    cmd = input('Enter do you want to continue? enter "Y" if yes:')
# When using with for opening a file no need to close the file.. 
# It atomatically closes the file after leaving the block.. 
```
