```python
dfTable[0] = dfTable[0].apply(lambda x:x.replace('\n',''))
dfTable[5] = dfTable[5].apply(lambda x:x.replace('\n',''))
dfTable[6] = dfTable[6].apply(lambda x:x.replace('\n',''))
dfTable[7] = dfTable[7].apply(lambda x:x.replace('\n',''))
```
